# extracting data

```python
month = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
year = str(2020)
day_in_month = [31,29,31,30,31,30,31,31,30,31,30,31]
data = []

for mon, days in tqdm(zip(month,day_in_month)):
    for day in range(1,days+1):
        day = str(day)
        driver.get('https://agmarknet.gov.in/SearchCmmMkt.aspx?Tx_Commodity=24&Tx_State=UP&Tx_District=1&Tx_Market=0&DateFrom='+day+'-'+mon+'-'+year+'&DateTo='+day+'-'+mon+'-'+year+'&Fr_Date='+day+'-'+mon+'-'+year+'&To_Date='+day+'-'+mon+'-'+year+'&Tx_Trend=0&Tx_CommodityHead=Potato&Tx_StateHead=Uttar+Pradesh&Tx_DistrictHead=Agra&Tx_MarketHead=--Select--')
        for tr in driver.find_elements_by_xpath('//*[@id="cphBody_GridPriceData"]/tbody//tr'):
            tds = tr.find_elements_by_tag_name('td')
            d = [td.text for td in tds]
            if d!=[]:
                data.append(d)
            # print ([td.text for td in tds])
            # break
        #break
    #break
```

# making dataframe from list and saving it.

```python
import pandas as pd
df = pd.DataFrame(data,columns=['Number','District Name','Market Name','Commodity','Variety','Grade','Min Price','Max Price','Model Price','Price Date'])
```
